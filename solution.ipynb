```python
import torchaudio
import audiocraft
from audiocraft.models import MusicGen
from IPython.display import Audio
from ipywidgets import Textarea, Button
import torch
```

```python
print(torch.cuda.is_available())
```

```python
model = MusicGen.get_pretrained('facebook/musicgen-small')
```

```python
model.set_generation_params(
    duration=15,        # 15 seconds of audio
    top_k=250,          # Standard diversity
    temperature=1.2,    # A bit more "creative/random"
    cfg_coef=4.5        # Stick closer to the text prompt
)
```

```python
result = model.generate(["Rock natural + nashed music to focus on learning"])
```

```python
Audio(result[0].cpu().numpy(), rate=model.sample_rate)
```

```python
# 1. Create the button widget
generate_tune_button = Button(
    description='Generate Tune', 
    button_style='success',
    tooltip='Click to generate music based on your description',
    icon='music'
)

# 2. Define the "Action" (What happens when clicked)
def start_generation(b):
    print("Generation started... this may take a moment.")
    
    # Move your generation logic INSIDE this function
    descriptions = ["calming natural music to focus on learning"]
    result = model.generate(descriptions)
    
    # Display the audio player specifically when the button is pressed
    display(Audio(result[0].cpu().numpy(), rate=model.sample_rate))
    print("Generation complete!")

# 3. LINK the button to the function
generate_tune_button.on_click(start_generation)

# 4. Display the button
display(generate_tune_button)
```
